#Cleaning our Dataset

##validity

```python
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping cities that are not Paris in the City attribute
# Explanation:Our analysis is only based on the city of Paris.
#
paris = df[df.City == 'Paris']
# previewing the first 10 rows 
paris.head(10)
# previewing the last 10 rows
#paris.tail(10)
```

```python
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping unnecessary attributes
# Explanation:We don't need them to answer any question.
#
# dropping month and year attribute
drop_monthyear = paris.drop(['month', 'year'], axis=1)
drop_monthyear.head(10)
# dropping Cars attribute
# we will use Bluecar counter attribute since they have common values
drop_cars =  drop_monthyear.drop('Cars', axis = 1)
drop_cars
```

```python
# Printing the shape of our new dataset
drop_cars.shape
# Printing the shape of our old dataset
#df.shape
```

```python
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping duplicates
#
drop_dupl = drop_cars.drop_duplicates()
drop_dupl
```

```python
# Printing the shape of our new dataset
drop_dupl.shape
```

```python
# Checking if there are missing values in your data
# counting the number of missing values
np.count_nonzero(drop_dupl.isnull())
```

```python
# Drop rows where all cells in that row is NAN
clean_rows = drop_dupl.dropna(how='all')
clean_rows
```

```python
# Drop columns where all cells in that column is NAN
clean_columns = clean_rows.dropna(axis=1, how='all')
clean_columns
```

```python
# Printing the shape of our new dataset
clean_columns.shape
```

```python
# Fill in missing data with zeros
clean_columns.fillna(0)
```

## Accuracy

```python
# Data Cleaning Action:
#Identifying and dropping outliers
#
import matplotlib.pyplot as plt 
% matplotlib inline 
#Calculating our first, third quantiles and then later our IQR
Q1 = clean_columns.quantile(0.25)
Q3 = clean_columns.quantile(0.75)
IQR = Q3 - Q1
#Removing outliers based on the IQR range and stores the result in the data frame 'df_out'
df_out = clean_columns[~((clean_columns < (Q1 - 1.5 * IQR)) | (clean_columns > (Q3 + 1.5 * IQR))).any(axis=1)]
# Printing the shape of our new dataset
print(df_out.shape)
#Printing the shape of our old dataset
print(clean_columns.shape)
#print(IQR)
#print(clean_columns < (Q1 - 1.5 * IQR)) |(clean_columns > (Q3 + 1.5 * IQR))
```
